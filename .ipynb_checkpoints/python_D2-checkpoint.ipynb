```python
df1=pd.DataFrame({'key':['x','y','z'],'lval':[1,2,99]})
```

```python
df2=pd.DataFrame({'key':['x','y','w'],'rval':[3,4,98]})
```

```python
pd.merge(df1,df2)  #inner join. finding common elements
```

```python
pd.merge(df1,df2, how="left") #left join
```

```python
pd.merge(df1,df2, how="right") #right join
```

```python
df1=pd.DataFrame({'key':['x','x','y','z'],'lval':[1,33,2,99]})
pd.merge(df1,df2, how="outer") #outer join
```

```python

df1=pd.DataFrame({'key1':['x','y','z'],'lval':[1,2,99]})
df2=pd.DataFrame({'key2':['x','y','z'],'rval':[3,4,98]})
pd.merge(df1,df2, how="left", left_on="key1", right_on="key2") #left join
```
